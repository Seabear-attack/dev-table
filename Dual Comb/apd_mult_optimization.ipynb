```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import Planck as h
from scipy.constants import elementary_charge as e
from scipy.constants import speed_of_light as c

def excess_noise(M, k):
    return M * k + (1-k) * (2-1/M)
photon_multiplier = np.logspace(0,2, 1000)
ion_ratio = 0.05

plt.semilogx(photon_multiplier, excess_noise(photon_multiplier, ion_ratio), label=f'k={ion_ratio}')
plt.xlabel('Multiplication factor M')
plt.ylabel('Excess noise F(M)')
print(excess_noise(50, ion_ratio)/ excess_noise(10, ion_ratio))

ion_ratio = 0.02

plt.semilogx(photon_multiplier, excess_noise(photon_multiplier, ion_ratio), label=f'k={ion_ratio}')
plt.xlabel('Multiplication factor M')
plt.ylabel('Excess noise F(M)')
plt.legend()
print(excess_noise(50, ion_ratio)/ excess_noise(10, ion_ratio))
```
